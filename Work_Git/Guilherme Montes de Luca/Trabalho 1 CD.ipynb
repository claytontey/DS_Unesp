# Calcular média, moda e mediana de pelo menos 2 variáveis quantitativas.

```python
sono_map = {
    "Menos de 4 horas": 3,
    "Entre 4 e 6 horas": 5,
    "Entre 6 e 8 horas": 7,
    "Mais de 8 horas": 9
}
descansado_map = {
    "Nunca": 1,
    "Raramente": 2,
    "Frequentemente": 3,
    "Sempre": 4
}

df["Horas_num"] = df["Horas de sono por noite"].map(sono_map)
df["Descansado_num"] = df["Sente-se descansado ao acordar"].map(descansado_map)

# Cálculo das estatísticas descritivas
estatisticas = {}

# Horas de sono por noite
estatisticas["Horas de sono por noite"] = {
    "Média": df["Horas_num"].mean(),
    "Moda": df["Horas_num"].mode()[0],
    "Mediana": df["Horas_num"].median()
}

# Sente-se descansado ao acordar
estatisticas["Sente-se descansado ao acordar"] = {
    "Média": df["Descansado_num"].mean(),
    "Moda": df["Descansado_num"].mode()[0],
    "Mediana": df["Descansado_num"].median()
}

# Exibir os resultados
for var, stats in estatisticas.items():
    print(f"\n📌 {var}")
    for stat_name, value in stats.items():
        print(f"{stat_name}: {value:.2f}")
```
