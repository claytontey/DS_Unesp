# Criando DataFrame do vinho vermelho

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lendo o dataset red
data_red = pd.read_csv(r"C:\Users\pirsp\Documents\Cincia_de_dados\P1\dataset\winequality-red.csv", sep=';')
df_red = pd.DataFrame(data_red)
print(df_red.to_string(index=False))
```

# Criando DataFrame do vinho branco

```python
# Lendo o dataset white
data_white = pd.read_csv(r"C:\Users\pirsp\Documents\Cincia_de_dados\P1\dataset\winequality-white.csv", sep=';')
df_white = pd.DataFrame(data_white)
print(df_white.to_string(index=False))
```

# Explorao dos dados

- Describe para uma vizualizao de analises mais basicas

```python
#Vinho vermelho
df_red.describe()
```

```python
#Vinho branco
df_white.describe()
```

- Anlise sobre a influencia da quantidade de aucar na qualidade do vinho

```python
# Tabela do vinho vermelho
red = df_red.groupby('quality')['residual sugar'].sum().reset_index()
print(red.sort_values(by='residual sugar', ascending=True).to_string(index=False))
```

```python
# Tabela do vinho branco
white = df_white.groupby('quality')['residual sugar'].sum().reset_index()
print(white.sort_values(by='residual sugar', ascending=True).to_string(index=False))
```

Nota-se que as maiores quantidades de aucar se concentram em classificaes de qualidade medianas

- Analise da influencia do ph na qualidade do vinho

```python
# Vinho Vermelho
red = df_red.groupby('quality')['pH'].mean().reset_index()
red['pH'] = red['pH'].round(2)
print(red.to_string(index=False))
```

Nota-se que em mdia o ph do vinho  3 e no tem influncia sobre sua qualidade

## Limpeza de Dados

- Retirar e substituir clulas Nan, caso existam

```python
# Vinho Vermelho
df_red.fillna(0, inplace=True)

#Vinho Branco
df_white.fillna(0, inplace=True)
```

- Arredondar os valores para duas casa decimais a fim de produzir uma melhor vizualizao dos dados

```python
# Vinho Vermelho
df_red = df_red.round(2)
df_red
```

```python
# Vinho Branco
df_white = df_white.round(2)
df_white
```

- Verificar se h dados faltantes

```python
print('Quantidade de dados faltantes:')

df_red.isnull().sum()
```

## Anlise Explortria

- O alcool est diretamente relacionado com a qualidade do vinho?

```python
# Vinho Vermelho
red = df_red.groupby('quality')['alcohol'].mean().reset_index()
print(red.sort_values(by='quality', ascending=True).to_string(index=False))
```

```python
#Vinho Branco
white = df_white.groupby('quality')['alcohol'].mean().reset_index()
print(white.sort_values(by='quality', ascending=True).to_string(index=False))
```

Com as informaes acima, pode-se dizer que concentraes de alcool maiores do que 11, em mdia, levam a uma melhor qualidade do vinho

- A acidez influncia no ph do vinho?

```python
# Vinho vermelho
red = df_red
red['pH'] = red['pH'].astype(int)
red = df_red.groupby('pH')['fixed acidity'].mean().reset_index()
print(red.sort_values(by='pH', ascending=True).to_string(index=False))
```

```python
# Vinho branco
white = df_white
white['pH'] = white['pH'].astype(int)
white = df_white.groupby('pH')['fixed acidity'].mean().reset_index()
print(white.sort_values(by='pH', ascending=True).to_string(index=False))
```

Nota-se que no vinho vermelho a acidez possui muito mais influencia sobre o ph do que no vinho branco

## Vizualizaes

- Utilizando heatmap primeiramente para tornar as relaes mais visuais

```python
#Vinho vermelho
sns.heatmap(df_red.corr())
```

```python
# Vinho Branco
sns.heatmap(df_white.corr())
```

- Grafico em linha mostrando o aumento da concentrao de alcool e o estado de qualidade do vinho

```python
#Vinho vermelho
df_alcool_quali = df_red.groupby('quality')['alcohol'].mean().reset_index()
df_alcool_quali = df_alcool_quali.sort_values(by='alcohol', ascending=True)
df_alcool_quali.plot(x='alcohol', y='quality', kind='line')
plt.title('Aumento da concentrao de alcool vs qualidade (vermelho)')
plt.xlabel('Concentrao de alcool')
plt.ylabel('Qualidade do vinho')
plt.show()
```

```python
#Vinho branco
df_alcool_quali = df_white.groupby('quality')['alcohol'].mean().reset_index()
df_alcool_quali = df_alcool_quali.sort_values(by='alcohol', ascending=True)
df_alcool_quali.plot(x='alcohol', y='quality', kind='line')
plt.title('Aumento da concentrao de alcool vs qualidade (branco)')
plt.xlabel('Concentrao de alcool')
plt.ylabel('Qualidade do vinho')
plt.show()
```

- Relao entre a densidade e o fixed acidity

```python
#Vinho vermelho
df_densi_fixed = df_red.groupby('density')['fixed acidity'].mean().reset_index()
df_densi_fixed = df_densi_fixed.sort_values(by='fixed acidity', ascending=True)
df_densi_fixed.plot(x='fixed acidity', y='density', kind='line')
plt.title('Aumento da acidez e a influencia na densidade (branco)')
plt.show()
```

```python
#Vinho branco
df_densi_fixed = df_white.groupby('density')['fixed acidity'].mean().reset_index()
df_densi_fixed = df_densi_fixed.sort_values(by='fixed acidity', ascending=True)
df_densi_fixed.plot(x='fixed acidity', y='density', kind='line')
plt.title('Aumento da acidez e a influencia na densidade (branco)')
plt.show()
```

Nota-se que no vinho branco a relao entre densidade e acidez  muito mais linear do que no vermelho
